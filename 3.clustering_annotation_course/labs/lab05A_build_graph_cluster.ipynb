# Lab 5A: Build Neighbor Graph + Cluster

## Objectives
- Build a kNN graph in PCA space
- Run Leiden clustering
- Save cluster assignments

## Outputs
- `../results/lab05A_clusters.csv`

---

Note: This lab is a focused version of `lab05_clustering.ipynb` (which explores multiple resolutions).


```python
import scanpy as sc
from pathlib import Path

adata = sc.datasets.pbmc3k_processed()

# Neighbor graph + clustering
sc.pp.neighbors(adata, n_neighbors=15, n_pcs=40)
sc.tl.leiden(adata, resolution=0.5, key_added='leiden')
sc.tl.umap(adata)

sc.pl.umap(adata, color='leiden', title='Leiden clusters (res=0.5)')

# Save clusters
Path('../results').mkdir(exist_ok=True)
out = Path('../results/lab05A_clusters.csv')
adata.obs[['leiden']].to_csv(out)
print(f"Wrote {out}")
```
